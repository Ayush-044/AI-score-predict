```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for col in ['Teacher_Quality', 
            'Parental_Education_Level', 
            'Distance_from_Home']:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_cols = [
    "Parental_Involvement",
    "Access_to_Resources",
    "Motivation_Level",
    "Family_Income",
    "Teacher_Quality",
    "Parental_Education_Level",
    "Peer_Influence"
]

categories = [
    ["Low", "Medium", "High"],
    ["Low", "Medium", "High"],
    ["Low", "Medium", "High"],
    ["Low", "Medium", "High"],
    ["Low", "Medium", "High"],
    ["High School", "College", "Postgraduate"],
    ["Positive","Neutral","Negative"]
]
```

```python
oe = OrdinalEncoder(categories=categories)

df[ordinal_cols] = oe.fit_transform(df[ordinal_cols])
```

```python
df[ordinal_cols].head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()


df["Internet_Access"] = le.fit_transform(df["Internet_Access"])
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
X = df.drop(columns=[
    "Exam_Score",
    "Gender",
    "Distance_from_Home",
    "School_Type",
    "Extracurricular_Activities",
    "Physical_Activity",
    "Learning_Disabilities"
])

y = df["Exam_Score"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(X_train.dtypes)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_unpruned = DecisionTreeRegressor(random_state=42)
dt_unpruned.fit(X_train, y_train)
```

```python
y_train_pred = dt_unpruned.predict(X_train)
y_test_pred = dt_unpruned.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

print("Unpruned Train R2:", r2_score(y_train, y_train_pred))
print("Unpruned Test R2 :", r2_score(y_test, y_test_pred))

print("Unpruned Train MAE:", mean_absolute_error(y_train, y_train_pred))
print("Unpruned Test MAE :", mean_absolute_error(y_test, y_test_pred))

print("Unpruned Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("Unpruned Test RMSE :", np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_pruned = DecisionTreeRegressor(
    max_depth=12,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
dt_pruned.fit(X_train, y_train)
```

```python
y_train_pred = dt_pruned.predict(X_train)
y_test_pred = dt_pruned.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

print("Pruned Train R2:", r2_score(y_train, y_train_pred))
print("Pruned Test R2 :", r2_score(y_test, y_test_pred))

print("Pruned Train MAE:", mean_absolute_error(y_train, y_train_pred))
print("Pruned Test MAE :", mean_absolute_error(y_test, y_test_pred))

print("Pruned Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("Pruned Test RMSE :", np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

```python
import joblib


joblib.dump(dt_pruned, "decision_tree_model.pkl")

joblib.dump(oe, "ordinal_encoder.pkl")


print("ML pipeline saved successfully!")
```
